```python
from stl import mesh
from mpl_toolkits import mplot3d
from matplotlib import pyplot
```

```python

STLFiles=["Needle_4cm.stl","Needle_9cm.stl","Needle_14cm.stl"]

needle=mesh.Mesh.from_file(STLFiles[0])
needle2=mesh.Mesh.from_file(STLFiles[1])
needle3=mesh.Mesh.from_file(STLFiles[2])
```

```python
# Create a new plot
figure = pyplot.figure(figsize=(15,15))
axes = figure.add_subplot(projection='3d')
percent=0.1
pointPercent=round(len(needle.vectors)*percent)
axes.add_collection3d(mplot3d.art3d.Poly3DCollection(needle.vectors[pointPercent:len(needle.vectors)]))
# Auto scale to the mesh size
scale = needle.points.flatten()
axes.auto_scale_xyz(scale, scale, scale)
pyplot.show()

figure = pyplot.figure(figsize=(15,15))
axes = figure.add_subplot(projection='3d')
pointPercent=round(len(needle2.vectors)*percent)
axes.add_collection3d(mplot3d.art3d.Poly3DCollection(needle2.vectors[pointPercent:len(needle2.vectors)]))
# Auto scale to the mesh size
scale = needle2.points.flatten()
axes.auto_scale_xyz(scale, scale, scale)
pyplot.show()

figure = pyplot.figure(figsize=(15,15))
axes = figure.add_subplot(projection='3d')
pointPercent=round(len(needle3.vectors)*percent)
axes.add_collection3d(mplot3d.art3d.Poly3DCollection(needle3.vectors[pointPercent:len(needle3.vectors)]))
# Auto scale to the mesh size
scale = needle3.points.flatten()
axes.auto_scale_xyz(scale, scale, scale)
pyplot.show()
# Show the plot to the screen


```
