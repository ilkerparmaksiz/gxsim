```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

```python
### Units for Garfield Gas Files ###
## Efield V/cm
## Velocity cm/ns
## DT,DL are sqrt(cm)

#Load 
path="Xe_100_10bar.json"
f=open(path)

data=json.load(f)
print(data.keys())

plt.figure(figsize=(8,15))

plt.subplot(3,1,1)
plt.title("Drift Velocity")

vd=np.array([data["electron_drift_velocity"]][0]) ## Convert from cm/us
Efield=np.array([data["electric_field"]][0])
plt.plot(Efield,vd*10) 
plt.xlabel("EField (V/cm)")
plt.ylabel("Drift Velocity (mm/us)")
plt.xlim(100,600)
plt.subplot(3,1,2)
plt.title("Longitutional Diffusion")
dl=np.array([data["electron_longitudinal_diffusion"]][0])
dl=(dl**2*(vd*1e6))/2 #cm^2/s

plt.plot(Efield,dl)
plt.xlim(100,600)
plt.ylabel("DL (cm^2/s)")
plt.subplot(3,1,3)
dt=np.array([data["electron_transversal_diffusion"]][0])
dt=(dt**2*(vd*1e6))/2 #cm/us

plt.title("Transverse Diffusion")
plt.plot(Efield,dt)
plt.ylabel("DT (cm^2/s)")
plt.xlim(100,600)
```
